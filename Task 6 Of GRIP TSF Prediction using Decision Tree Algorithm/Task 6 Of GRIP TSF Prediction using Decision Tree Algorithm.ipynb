### Task 6: Prediction using Decision Tree Algorithm

Title: Perform 'Explore Decision Tree Algorithm' 

Objective: Create the Decision Tree classifier and visualize it graphically.
    
Tools Used: Python, with libraries like numpy, pandas, seaborn, matplotlib

Dataset: We use a dataset available at https://bit.ly/3kXTdox
        
The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

### Importing libraries

```python
import sklearn.datasets as datasets
import pandas as pd
```

### importing data sets

```python
iris=datasets.load_iris()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
X.head()
```

```python
X.tail()
```

```python
X.info()
```

```python
X.describe()
```

```python
X.isnull().sum()
```

```python
Y = iris.target
Y
```

### split dataset into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

### Defining the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)

print('Decision Tree Classifer Created Successfully')
```

```python
y_predict = dtc.predict(X_test)
```

### Constructing confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

### visualizing the Decision tree

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(dtc, feature_names = fn, class_names = cn, filled = True);
```
